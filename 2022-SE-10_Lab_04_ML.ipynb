```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\PMLS\\1.python\\Dataset .csv')
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["Aggregate rating"], bins=20, kde=True, color="blue")
plt.title("Distribution of Aggregate Ratings", fontsize=14, weight="bold")
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Count of Restaurants by Rating Text
plt.figure(figsize=(10, 5))
sns.countplot(x="Rating text", data=df, palette="viridis", order=df["Rating text"].value_counts().index)
plt.title("Count of Restaurants by Rating Text", fontsize=14, weight="bold")
plt.xlabel("Rating Text")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Price Range vs Aggregate Rating
plt.figure(figsize=(10, 5))
sns.boxplot(x="Price range", y="Aggregate rating", data=df, palette="coolwarm")
plt.title("Price Range vs Aggregate Rating", fontsize=14, weight="bold")
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.show()
```

```python
# 4. Top 10 Cities with Most Restaurants
plt.figure(figsize=(12, 6))
top_cities = df["City"].value_counts().nlargest(10)
sns.barplot(x=top_cities.index, y=top_cities.values, palette="magma")
plt.title("Top 10 Cities with the Most Restaurants", fontsize=14, weight="bold")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Online Delivery vs Aggregate Rating
plt.figure(figsize=(10, 5))
sns.boxplot(x="Has Online delivery", y="Aggregate rating", data=df, palette="Set2")
plt.title("Online Delivery vs Aggregate Rating", fontsize=14, weight="bold")
plt.xlabel("Has Online Delivery")
plt.ylabel("Aggregate Rating")
plt.show()
```

```python
Here are 10 key insights from the data visualization:

1. Most restaurants have average ratings between 3.0 and 4.5, with fewer 
restaurants rated below 2.5 or above 4.8.

2. "Very Good" and "Excellent" ratings dominate the dataset,
indicating a generally high customer satisfaction level.

3. Higher price range restaurants tend to have better ratings,
suggesting that customers are willing to pay more for better quality.

4. The top 10 cities with the most restaurants include large metropolitan areas,
showing that restaurant businesses are concentrated in urban hubs.

5. Restaurants offering online delivery tend to have slightly lower ratings 
compared to those without online delivery, possibly due to service or food quality issues.

6. There is a significant number of budget-friendly restaurants
(Price Range 1 and 2), but their average ratings vary, showing mixed customer satisfaction.

7. The distribution of aggregate ratings shows a peak around 3.5 to 4.5,
indicating that most restaurants maintain a fairly good quality.

8. Cities with more restaurants likely have more diverse cuisine options,
which could attract a wider customer base.

9. Ratings are not evenly distributed across all price ranges,
with mid-to-high-range restaurants receiving better ratings on average.

10. Restaurants with table booking generally have higher ratings,
which could indicate a better dining experience.
```
